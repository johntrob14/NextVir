```python
import matplotlib.pyplot as plt
import numpy as np

main_tprs = {}
main_fprs = {}
main_thresh = {}
prefixes = ['NextVir', 'dvf', 'virt']
labels = ['NextVir', 'DeepVirFinder', 'Virtifier', 'XVir']
colors = ['b', 'g', 'r', 'orange']
for i in range(len(prefixes)):
    prefix = './curves/' + prefixes[i]
    main_tprs[labels[i]] = np.load(prefix + '_main_tpr.npy')
    main_fprs[labels[i]] = np.load(prefix + '_main_fpr.npy')
    main_thresh[labels[i]] = np.load(prefix + '_main_thresh.npy')
main_tprs['XVir'] = np.load('./curves/xvir_icav_roc.npz')['tpr']
main_fprs['XVir'] = np.load('./curves/xvir_icav_roc.npz')['fpr']
plt.figure()
# for i in range(len(main_tprs)):
for i in ['DeepVirFinder', 'NextVir', 'Virtifier', 'XVir']:
    plt.plot(main_fprs[i], main_tprs[i], label=f'data %s' % i, color=colors[labels.index(i)])

sum = 0
last = 0
for i in range(len(main_tprs['NextVir'])):
    sum += main_tprs['NextVir'][i] * (main_fprs['NextVir'][i] - last)
    last = main_fprs['NextVir'][i]
print(sum)
sum = 0
last = 0
for i in range(len(main_tprs['DeepVirFinder'])):
    sum += main_tprs['DeepVirFinder'][i] * (main_fprs['DeepVirFinder'][i] - last)
    last = main_fprs['DeepVirFinder'][i]
print(sum)

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```
